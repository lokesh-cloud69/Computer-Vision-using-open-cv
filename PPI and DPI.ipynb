```python
import cv2

# Read image
img = cv2.imread('butterfly.jpg')

# Get dimensions
height, width = img.shape[:2]

# Assume image width in inches
width_inches = 8

ppi = width / width_inches

print("Image Width:", width, "pixels")
print("Width in Inches:", width_inches)
print("PPI =", round(ppi, 2))
```

```python
import cv2
import math

img = cv2.imread('butterfly.jpg')

height, width = img.shape[:2]

# Physical dimensions
width_inches = 8
height_inches = 6

ppi_width = width / width_inches
ppi_height = height / height_inches

print("Horizontal PPI:", round(ppi_width, 2))
print("Vertical PPI:", round(ppi_height, 2))
```

```python
import cv2
import math

img = cv2.imread('butterfly.jpg')

height, width = img.shape[:2]

screen_size = 15.6   # inches

diagonal_pixels = math.sqrt(
    width**2 + height**2
)

ppi = diagonal_pixels / screen_size

print("Diagonal Pixels:", round(diagonal_pixels, 2))
print("PPI:", round(ppi, 2))
```

```python
import cv2

img = cv2.imread('butterfly.jpg')

height, width = img.shape[:2]

print_width_inches = 5

dpi = width / print_width_inches

print("Image Width:", width)
print("Print Width:", print_width_inches)
print("DPI:", round(dpi, 2))
```
