```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

rows, cols = img.shape[:2]

# Shift 100 pixels right and 50 pixels down
M = np.float32([
    [1, 0, 100],
    [0, 1, 50]
])

translated = cv2.warpAffine(img, M, (cols, rows))

result = np.hstack((img, translated))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Resize to specific dimensions
resized = cv2.resize(img, (300, 300))

print("Original Shape:", img.shape)
print("Resized Shape:", resized.shape)

# To use hstack, heights must match
original_resized = cv2.resize(img, (300, 300))

result = np.hstack((original_resized, resized))

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Resized")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

rows, cols = img.shape[:2]

M = cv2.getRotationMatrix2D(
    (cols/2, rows/2),
    45,
    1
)

rotated = cv2.warpAffine(
    img,
    M,
    (cols, rows)
)

result = np.hstack((img, rotated))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

horizontal = cv2.flip(img, 1)
vertical = cv2.flip(img, 0)
both = cv2.flip(img,-1)

result = np.hstack((img, horizontal, vertical,both))

plt.figure(figsize=(15,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

rows, cols = img.shape[:2]

pts1 = np.float32([
    [50,50],
    [200,50],
    [50,200]
])

pts2 = np.float32([
    [10,100],
    [200,50],
    [100,250]
])

M = cv2.getAffineTransform(pts1, pts2)

affine = cv2.warpAffine(
    img,
    M,
    (cols, rows)
)

result = np.hstack((img, affine))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

rows, cols = img.shape[:2]

pts1 = np.float32([
    [50,50],
    [cols-50,50],
    [50,rows-50],
    [cols-50,rows-50]
])

pts2 = np.float32([
    [0,0],
    [cols,0],
    [100,rows],
    [cols-100,rows]
])

M = cv2.getPerspectiveTransform(
    pts1,
    pts2
)

perspective = cv2.warpPerspective(
    img,
    M,
    (cols, rows)
)

result = np.hstack((img, perspective))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
