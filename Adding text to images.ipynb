```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Add text
cv2.putText(
    img,
    'OpenCV',
    (50, 100),
    cv2.FONT_HERSHEY_SIMPLEX,
    2,
    (0, 255, 0),
    3
)

# Display image
plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Image with Text")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Create copy
text_img = img.copy()

# Add text
cv2.putText(
    text_img,
    'Butterfly',
    (40, 80),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.5,
    (255, 0, 0),
    3
)

# Display side by side
result = np.hstack((img, text_img))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Text Added")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

cv2.putText(img, "OpenCV", (30,80),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (0,255,0), 2)

cv2.putText(img, "Image Processing", (30,150),
            cv2.FONT_HERSHEY_COMPLEX,
            1, (255,0,0), 2)

plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
