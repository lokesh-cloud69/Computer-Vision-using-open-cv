```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read grayscale image
img = cv2.imread('butterfly.jpg',0)

# Create kernel
kernel = np.ones((5,5), np.uint8)

# Erosion
eroded = cv2.erode(
    img,
    kernel,
    iterations=1
)

# Display
result = np.hstack((img, eroded))

plt.figure(figsize=(12,5))
plt.imshow(result, cmap='gray')
plt.title("Original | Eroded")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg', 0)

# Kernel
kernel = np.ones((5,5), np.uint8)

# Morphological operations
eroded = cv2.erode(img, kernel, iterations=1)
dilated = cv2.dilate(img, kernel, iterations=1)

# Stack images
result = np.hstack((
    img,
    eroded,
    dilated
))

plt.figure(figsize=(15,5))
plt.imshow(result, cmap='gray')
plt.title("Original | Erosion | Dilation")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg', 0)

kernel3 = np.ones((3,3), np.uint8)
kernel7 = np.ones((7,7), np.uint8)

erode3 = cv2.erode(img, kernel3)
erode7 = cv2.erode(img, kernel7)

result = np.hstack((
    img,
    erode3,
    erode7
))

plt.figure(figsize=(15,5))
plt.imshow(result, cmap='gray')
plt.title("Original | 3x3 Erosion | 7x7 Erosion")
plt.axis('off')
plt.show()
```

```python
#Import package
import cv2 
from pylab import *
from skimage import io
#Read image 
image = cv2.imread("butterfly.jpg") 
#Define erosion size 
s1 = 0
s2 = 10
s3 = 10
#Define erosion type 
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
#Define and save the erosion template 
tmp1 = cv2.getStructuringElement(t1, (2*s1 + 1, 2*s1+1), (s1, s1))
tmp2= cv2.getStructuringElement(t2, (2*s2 + 1, 2*s2+1), (s2, s2))
tmp3 = cv2.getStructuringElement(t3, (2*s3 + 1, 2*s3+1), (s3, s3))
#Apply the erosion template to the image and save in different variables
final1 = cv2.erode(image, tmp1) 
final2 = cv2.erode(image, tmp2) 
final3 = cv2.erode(image, tmp3)
#Show all the images with different erosions
figure(0) 
io.imshow(final1)
figure(1) 
io.imshow(final2)
figure(2) 
io.imshow(final3)
```

```python
#Import package
import cv2 
from pylab import *
from skimage import io
#Read image 
image = cv2.imread("butterfly.jpg") 
#Define dilation size 
d1 = 0
d2 = 10
d3 = 10
#Define dilation type 
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
#Define and save the dilation template 
tmp1 = cv2.getStructuringElement(t1, (2*d1 + 1, 2*d1+1), (d1, d1))
tmp2= cv2.getStructuringElement(t2, (2*d2 + 1, 2*d2+1), (d2, d2))
tmp3 = cv2.getStructuringElement(t3, (2*d3 + 1, 2*d3+1), (d3, d3))
#Apply the dilation template to the image and save in different variables
final1 = cv2.dilate(image, tmp1) 
final2 = cv2.dilate(image, tmp2) 
final3 = cv2.dilate(image, tmp3)
#Show all the images with different dilations
figure(0) 
io.imshow(final1)
figure(1) 
io.imshow(final2)
figure(2) 
io.imshow(final3)
```
