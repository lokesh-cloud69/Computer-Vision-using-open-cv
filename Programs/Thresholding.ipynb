```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg', 0)

ret, binary = cv2.threshold(
    img,
    127,
    255,
    0
)

result = np.hstack((img, binary))

plt.figure(figsize=(12,5))
plt.imshow(result, cmap='gray')
plt.title("Original | Binary")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg', 0)

_, t0 = cv2.threshold(img, 127, 255, 0)
_, t1 = cv2.threshold(img, 127, 255, 1)
_, t2 = cv2.threshold(img, 127, 255, 2)
_, t3 = cv2.threshold(img, 127, 255, 3)
_, t4 = cv2.threshold(img, 127, 255, 4)

result = np.hstack((img, t0, t1, t2, t3, t4))

plt.figure(figsize=(20,5))
plt.imshow(result, cmap='gray')
plt.title("Original | 0 | 1 | 2 | 3 | 4")
plt.axis('off')
plt.show()
```
