```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Average smoothing
avg = cv2.blur(img, (15,15))

# Display
result = np.hstack((img, avg))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Average Blur")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

gaussian = cv2.GaussianBlur(
    img,
    (15,15),
    0
)

result = np.hstack((img, gaussian))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

bilateral = cv2.bilateralFilter(
    img,
    9,
    75,
    75
)

result = np.hstack((img, bilateral))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Bilateral Filter")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

average = cv2.blur(img, (15,15))

gaussian = cv2.GaussianBlur(
    img,
    (15,15),
    0
)

median = cv2.medianBlur(
    img,
    15
)

bilateral = cv2.bilateralFilter(
    img,
    9,
    75,
    75
)

# Stack images
result = np.hstack((
    img,
    average,
    gaussian,
    median,
    bilateral
))

plt.figure(figsize=(20,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title(
    "Original | Average | Gaussian | Median | Bilateral"
)
plt.axis('off')
plt.show()
```
